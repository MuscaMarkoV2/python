Wine Quality Data Set - contains 12 parameters which characterise the wine:
fixed acidity
volatile acidity
citric acid
residual sugar
parametrul chlorides
free sulfur dioxide
total sulfur dioxide
density
pH
sulphates
alcohol
quality (values between 0 and 10) - parameter to determinate
The first 11 parameters can be used to classify wines into 11 categories, the category indicated by the quality parameter.
For example, if the quality parameter takes the value 3, then wines with this quality are part of class 3.


```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("wine.csv")
```

```python
valori = data.values
print(valori)
```

```python
x = valori[:, 0:11]
y = valori[:, 11]
print(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2)
model = LogisticRegression()
```

```python
# 8. AntrenÄƒm modelul
history = model.fit(
    x_train, y_train
)

print("\nðŸ”” Antrenament terminat!")
```

```python
etichete_prezise = model.predict(x_test)
```

```python
acuratete = model.score(x_test, y_test)
print(acuratete)
```

```python
for i, et in enumerate(etichete_prezise):
    print(f"Eticheta reala:{y_test[i]}, eticheta prezisa:{et}")
```
