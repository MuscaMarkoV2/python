Setul de date al calitatii vinului (Wine Quality Data Set)

contine 12 parametrii care caracterizeaza vinul:
1 - aciditatea fixa (parametrul fixed acidity)

2 - aciditatea volatila (parametrul volatile acidity)

3 - acidul citric (parametrul citric acid)

4 - zaharul rezidual (parametrul residual sugar)

5 - cloruri (parametrul chlorides)

6 - dioxid de sulf liber (parametrul free sulfur dioxide)

7 - dioxid de sulf total(parametrul total sulfur dioxide)

8 - densitate (parametrul density)

9 - pH (parametrul pH)

10 - sulfaÈ›i (parametrul sulphates)

11 - alcool (parametrul alcohol)

12 - calitate (parametrul quality) (valori intre 0 si 10) - parametrul ce se doreste determinat

Primii 11 parametrii pot fi utilizati pentru clasificarea vinurilor in 11 categorii, categorie indicata de parametrul calitate. De exemplu daca parametrul calitate ia valoarea 3, atunci vinurile cu aceasta calitate fac parte din clasa 3.

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("vin.csv")
```

```python
valori = data.values
print(valori)
```

```python
x = valori[:, 0:11]
y = valori[:, 11]
print(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2)
model = LogisticRegression()
```

```python
# 8. AntrenÄƒm modelul
history = model.fit(
    x_train, y_train
)

print("\nðŸ”” Antrenament terminat!")
```

```python
etichete_prezise = model.predict(x_test)
```

```python
acuratete = model.score(x_test, y_test)
print(acuratete)
```

```python
for i, et in enumerate(etichete_prezise):
    print(f"Eticheta reala:{y_test[i]}, eticheta prezisa:{et}")
```
